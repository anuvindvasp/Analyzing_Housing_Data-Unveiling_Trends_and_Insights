```python
# Importing module from python library

import matplotlib.pyplot as plt
import seaborn as sns
```

## UNIVARIATE ANALYSIS

```python
# Boxplot

#Boxplot of price
sns.boxplot(data=df, y='price')
plt.title('Boxplot of Price')
plt.ylabel('Price')
plt.show()
```

```python
# outliers based on IQR for the 'price' column

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
print("IQR : ",IQR)
print("Lower Bound : ",lower_bound)
print("Upper Bound : ",upper_bound)
```

```python
# To remove outliers
filtered_df= df['price'][(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
filtered_df
sns.boxplot(y=filtered_df)
plt.title('Boxplot after handling outliers')
plt.xlabel('price')
plt.show()
```

```python
## Box plot for price by number of bedrooms

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='bedrooms', y='price')
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
## Box plot for price by number of bedrooms after removing outliers

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_no_outliers, x='bedrooms', y='price')
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Pie Chart

##Condition of House
conditions = df['condition'].value_counts()
plt.pie(conditions, labels=conditions.index, autopct='%1.2f%%')
plt.title('Pie Chart')
plt.legend(loc='upper right')
plt.show()
```

```python
# Countplot

# Prepare the data in a format suitable for a single count plot
melted_df = pd.melt(df[['bedrooms', 'bathrooms', 'floors']],var_name='Feature',value_name='Count')

# Plot the count plot
plt.figure(figsize=(9,6))
sns.countplot(data=melted_df, x='Count', hue='Feature', palette='Set1')

# Add title and labels
plt.title('Distribution of Bedrooms, Bathrooms, and Floors', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Feature', fontsize=10, title_fontsize=12)
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```

**1. Distribution of Bedrooms:**

Most properties likely have 2–4 bedrooms.

The most frequent value is probably 3 bedrooms, as this is typical for family homes.

Outliers : Houses with 1 bedroom (studio or apartments) or more than 6 bedrooms (luxury or multi-family homes) are rare.

Insight:
The majority of houses cater to medium-sized families.
Properties with more bedrooms could be marketed as premium or multi-family homes.

**2. Distribution of Bathrooms:**

Most properties likely have 1–3 bathrooms.

The most frequent value might be 2 bathrooms, which is typical for single-family homes.

Outliers : Properties with fewer than 1 bathroom are rare.
Houses with more than 4 bathrooms likely belong to luxury or high-end properties.

Insight:
Bathrooms tend to scale with bedrooms, but there are more homes with fewer bathrooms than bedrooms.
Buyers looking for affordability may favor homes with fewer bathrooms.

**3. Distribution of Floors:**

Most houses likely have 1 or 2 floors.

Single-floor houses are likely the most common.

Outliers:
Houses with more than 2 floors (3 or 4 floors) are rare and usually luxury properties or townhouses.

Insight:
Single-story homes are ideal for retirees or those preferring accessibility.
Multi-story homes might appeal to larger families or those wanting more square footage without increasing the lot size.

```python
#Lineplot

# For 2 numerical variables

# Price vs Squarefeet_living
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='sqft_living', y='price')
plt.title('Price vs Sqrft living')
plt.show()
```

```python
# Barplot

# For One Categorical and one numerical Variable:

# Calculate the average price by city
city_prices = df.groupby('city')['price'].mean().sort_values()

# Plot the data as a bar chart
plt.figure(figsize=(15, 7))
city_prices.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Housing Price by City', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.tight_layout()
plt.show()
```



*   Medina city has more pricing for housing as compared to othrt cities



```python
# Clustered Bar Plot

# For 2 categorical variables

# Calculate counts for each combination
count_df = df.groupby(['city', 'condition']).size().reset_index(name='price')

# Pivot the dataframe for easy plotting
pivot_df = count_df.pivot(index='city', columns='condition', values='price')

# Plot clustered bar plot
plt.figure(figsize=(10,5))
pivot_df.plot(kind='bar', figsize=(14,8),width=1.3)
plt.title('Clustered Bar Plot of City vs Condition')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend(title='Condition', fontsize=10)
plt.grid(axis='x', alpha=0.7)
plt.tight_layout()
plt.show()
```

## FEATURE ENGINEERING

```python
# Add a column for the age of the property

import numpy as np
current_year = 2025  # Update this to the current year
df['property_age'] = current_year - df['yr_built']

df['property_age'] = np.where(df['yr_renovated'] > 0,current_year - df['yr_renovated'],df['property_age'])          # Handle properties with renovations

df[['property_age']].head()    # Display the first few rows to verify the new columns
```

```python
# Add a column for price per square foot
df['price_per_sqft'] = df['price'] / df['sqft_living']

df[['price_per_sqft']].head()         # Display the first few rows to verify the new columns
```

```python
# Add column for Total rooms
df['total_rooms'] = df['bedrooms'] + df['bathrooms']
df[['total_rooms']].head()
```

```python
df = df.drop(columns=['country'], errors='ignore')
df.head()
```
